This json dataset records China's GDP from 1960-2016. This part of the homework, while unfinished, is rather clear-cut in the information that it presents, but the cleaning process required a looped dictionary redefinition to parse out the unneeded data. 

```python
import pandas as pd
```

```python
import json
```

```python
import requests
```

```python
get_request = requests.get('http://api.worldbank.org/countries/CHN/indicators/NY.GDP.MKTP.CD?per_page=5000&format=json')
```

Now that we have imported the data, this is what it currently looks like:

```python
get_request.text
```

```python
tariffs_df = json.loads(get_request.text)
```

And this is what it looks like in a more presentable format:

```python
print (json.dumps(tariffs_df, indent=4, sort_keys=True))
```

However, the majority of these tuples are not pertinent to any coherent data analysis. In fact, we only require the year and the GDP; the remaining pairs are not of interest.

```python
actualdata = tariffs_df[1]
```

```python
actualdata[1].keys()
```

```python
gdp_dict = {} 
gdp_dict['Year'] = 'GDP (USD)',
for item in actualdata:
    if item['value'] is not None:
        gdp_dict[item['date']] = item['value']
```

```python
series = pd.Series(gdp_dict)
```
