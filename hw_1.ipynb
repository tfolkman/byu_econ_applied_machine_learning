## Part 2: Text File

For my text file, I found a list of frequently occurring last names from the 1990 census. The dataset lists surnames from most common to least common, along with each name's percentage frequency in the census and the cumulative frequency of that name and all names above it in the list. Looking at this dataset should show us which surnames were most and least common in 1990.

### Sub-part 1: Read in the data

```python
import pandas as pd
import numpy as np
census_names = pd.read_fwf('https://www2.census.gov/topics/genealogy/1990surnames/dist.all.last', header = None)
census_names.columns = ['surname', 'freq', 'cum_freq', 'rank']
```

```python
census_names.head()
```

```python
census_names.info()
```

### Sub-part 2: Clean the data

By looking at "census_names.info()" above, I can tell that we don't have any columns with all missing data, and we only have two rows out of nearly 89,000 that have missing data. I'll drop those two and move on to looking for outliers.

```python
census_names.dropna()
```

```python
%matplotlib inline
bbox = census_names['cum_freq'].plot(kind="box")
```

```python
plt.boxplot(census_names['freq'])
```

These boxplots don't seem to be showing any outliers (and I'm not sure how to scale the second one), but they do show the general distribution of the data. Name frequency and cumulative frequency are highly skewed, so by the time we reach the median name, we've already seen about 90% of the people in our sample.

### Sub-part 3: Describe the data

```python
census_names.describe()
```

Looking at the summary statistics for name frequency, we can see that more than 75% of last names recorded on the US census occur less than 0.000001% of the time in the data. I'll add on some more percentiles to see where the last names start to become visibly more common.

```python
census_names.describe(percentiles=[.01,.05,.75,.775,.80,.90,.95,.99])
```

There are a couple of interesting things here: first, the mean last name in the data occurs 0.000896% of the time in the 1990 census, but the median last name occurs less than 0.000001% of the time. A last name can be in the top 20% of most common names on the census if it occurs only 0.001% of the time in the data (or in other words, if it appears once in every 100,000 people). And finally, a last name only has to occur 0.013% of the time to be in top 1% on most common names.

```python
census_names.head(20)
```

As you'd expect, "Smith" and "Johnson" are the most common names on the 1990 census. At the same time, though, I was surprised that the first stereotypically "non-white" name was "Garcia" with a rank of 18. This suggests that although America is growing more diverse (or so I often hear), this country has a long way to go before "non-white" surnames dominate the list.

### Sub-part 4: Visualize the data

```python
sns.distplot(census_names.freq, axlabel = "Name Frequency (%)", label = "Distribution of Surnames on the 1990 Census")
```

```python
ax = sns.violinplot(x="freq", data=census_names)
```

These graphs show the general skewed pattern I described previously. In the future, I would re-scale the data to be able to see the distribution in more detail.

## Part 3: DTA File

For my work as an RA for Dr. Price, I manage a team of data trainers who use Stata to create linkages between census and vital records (this project is called LIFE-M). The data is divided into sets, where each set consists of a "primary" taken from, say, a birth record in 1850 and up to 30 "potentials" (e.g. death records that have the highest probability of matching with the primary). The data trainer's role is to decide which of those potentials is truly the same person as the one listed in the primary. 

Part of my role in managing the team is helping to keep track of the trainers' performance. The administrative team primarily cares about two metrics when judging an RA's performance: (1) their speed, measured by time spent on a batch of 40 sets, and (2) their accuracy, measured by percent agreement with a key on select "audit" batches.

In order to compare RAs' performance over time, I will use a dataset of all of the batches the LIFE-M team has trained since the beginning of the project a few years ago (each observation signifies an individual batch). To make the data anonymous, I first worked in Stata to randomly generate a unique identifier for each data trainer, then dropped the RAs' names from the dataset.
